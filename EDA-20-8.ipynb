```python
#printing null values columns in the form of list
l1=[]
l2=[]
d1=dict(athletes.isnull().any())
for i,j in d1.items():
    if j==False:
        l1.append(i)
    else:
        l2.append(i)
print("Null value columns are :")        
print(l2)
```

```python
#Total medal
athletes.Medal.value_counts()
```

```python
#Total no. of female athletes in each summer olympic
female_participents=athletes[(athletes.Sex=='F') & (athletes.Season=='Summer')][['Sex','Year']]
female_participents=female_participents.groupby('Year').count().reset_index()
female_participents
```

```python
#Total no. of participents in winter olympic by year
f_participents=athletes[(athletes.Sex=='F') & (athletes.Season=='Winter')][['Sex','Year']]
f_participents=f_participents.groupby('Year').count().reset_index()
f_participents
```

```python
womenOlympics=athletes[(athletes.Sex=='F') & (athletes.Season=='Summer')]
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(15,10))
sns.countplot(x='Year',data=womenOlympics,palette='Spectral')
plt.title('Women participents')
```
